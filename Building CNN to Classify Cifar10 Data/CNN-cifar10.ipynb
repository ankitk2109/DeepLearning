# Deep Learning with Keras: cifar10 datasets

## Importing Libraries 

```python
import numpy as np
np.random.seed(1337)  # for reproducibility
import pandas as pd

import matplotlib.pyplot as plt
import pydot
%matplotlib inline

import random
import scipy as sp
import PIL

from IPython.display import display, HTML, Image
from IPython.display import SVG

from TAS_Python_Utilities import data_viz
from TAS_Python_Utilities import visualize_tree

from sklearn import preprocessing 
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz

import keras
from keras.utils import to_categorical 
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import np_utils
from keras.utils.vis_utils import model_to_dot
from keras.datasets import cifar10
keras.backend.set_image_data_format('channels_last')
```

## 1. Loading the Dataset and Examining some examples

```python
# loading dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# one hot encode target values
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# summarizing the dataset
print('Train: X=%s, y=%s' % (X_train.shape,y_train.shape))
print('Test: X=%s, y=%s' % (X_test.shape, y_test.shape))
print('Train: y={0}'.format(y_train.shape))
print('Test: y={0}'.format(y_test.shape))
```

```python
# ploting first few images
for i in range(9):
    # define subplot
    plt.subplot(330 + 1 + i) #3:rows, 3:columns, 0+1+i:index
    # plot raw pixel data
    plt.imshow(X_train[i])
# show the figure

plt.show()
```

## 2. Building simple feedforward neural network without Convolution

```python
# Flattening input images into 3072 features
X_train_flat = X_train.reshape(X_train.shape[0],3072).astype('float32') #np.reshape(row,column)
X_test_flat = X_test.reshape(X_test.shape[0],3072).astype('float32')

# Scaling the flattened features
X_train_flat /= 255
X_test_flat /= 255
#print(X_train_flat.shape)

#Creating simple feedforward neural network without convolution.
m = Sequential()
m.add(Dense(input_dim = 3072, units= 32))
m.add(Dense(output_dim = 128, activation = 'relu'))
m.add(Dense(output_dim = 10, activation = 'softmax'))
m.summary()

#Compiling the model.
m.compile(optimizer='rmsprop' , loss='categorical_crossentropy', metrics=['accuracy'])

#Evaluating the performance
history = m.fit(X_train_flat, np.asfarray(y_train),
              batch_size=32,
              epochs=10,
              validation_data=(X_test_flat, np.asfarray(y_test)))
```

## 3. Evaluating the model on test and train

```python
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='lower right')
plt.ylim((0, 1))
plt.show()
```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
#plt.ylim((-0.1, 2))
plt.show()
```

```python
print("******** Training Data ********")
# Make a set of predictions for the training data
y_pred = m.predict_classes(np.asfarray(X_train_flat))


#We need non one-hot encoded values to compare in confusion matrix
y_train_rounded = np.argmax(y_train, axis=1)

# Print performance details
print(metrics.classification_report(y_train_rounded, y_pred))

# Print confusion matrix
print("Confusion Matrix")
display(pd.crosstab(y_train_rounded, y_pred, rownames=['True'], colnames=['Predicted'], margins=True))

print("****** Test Data ********")

# Make a set of predictions for the validation data
y_pred = m.predict_classes(np.asfarray(X_test_flat))
#print(y_pred[0])

#We need non one-hot encoded values to compare using confusion matrix
y_test_rounded = np.argmax(y_test, axis=1)
#print(rounded_test_labels[0])

# Print performance details
print(metrics.classification_report(y_test_rounded, y_pred))

# Print confusion matrix
print("Confusion Matrix")
display(pd.crosstab(y_test_rounded, y_pred, rownames=['True'], colnames=['Predicted'], margins=True))
```

## 5. Building a CNN

```python
#Building CNN model
model = Sequential()
#model.add(conv2D(numOfConvoluatedLayersRequired(32), filterSize=(3,3), inputShapeOfImage= (32, 32, 3), activation = 'relu'))
model.add(Conv2D(32, (3, 3) , input_shape= (32, 32, 3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Flatten())
model.add(Dense(output_dim = 128, activation = 'relu'))
model.add(Dense(output_dim = 10, activation = 'softmax'))
model.summary()

#Compiling the model.
model.compile(optimizer='rmsprop' , loss='categorical_crossentropy', metrics=['accuracy'])

#Scaling the training featues
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train = X_train / 255
X_test = X_test / 255

history = model.fit(X_train, y_train,
              batch_size=32,
              epochs=3,
              validation_data=(X_test, y_test))
```
